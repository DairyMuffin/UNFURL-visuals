```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 500)
# Two resonant waves in phase
y1 = np.sin(x)
y2 = np.sin(x + 0.2)  # Slightly phase-shifted
y_sum = y1 + y2

plt.figure(figsize=(8, 3))
plt.plot(x, y1, label='Resonant State 1')
plt.plot(x, y2, label='Resonant State 2')
plt.plot(x, y_sum, label='Combined Coherence', linewidth=2)
plt.title('Resonant States and Coherence')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 3))
plt.plot(x, np.sin(x), label='Resonant State')
plt.axhline(y=0.8, color='r', linestyle='--', label='Dissonance Threshold')
plt.fill_between(x, 0.8, 1.5, color='red', alpha=0.2, label='Yield Trigger Zone')
plt.title('Dissonance Triggering Yield')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.patches as patches

fig, ax = plt.subplots(figsize=(6,6))

circle = patches.Circle((0.5, 0.5), 0.4, fill=False, linewidth=2)
ax.add_patch(circle)

ax.arrow(0.5, 0.1, 0, 0.3, head_width=0.05, head_length=0.05, fc='blue', ec='blue')
ax.arrow(0.85, 0.5, -0.3, 0, head_width=0.05, head_length=0.05, fc='blue', ec='blue')
ax.arrow(0.5, 0.9, 0, -0.3, head_width=0.05, head_length=0.05, fc='blue', ec='blue')
ax.arrow(0.15, 0.5, 0.3, 0, head_width=0.05, head_length=0.05, fc='blue', ec='blue')

ax.text(0.5, 0.55, 'Coherence\nReinforcement', ha='center', va='center')
ax.text(0.5, 0.02, 'Feedback Loop', ha='center')
ax.axis('off')
plt.title('Recursive Feedback Loop')
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8,4))

# Resonant Conjunction (AND)
ax.plot([0.1,0.3], [0.7,0.7], 'k-', linewidth=2)
ax.plot([0.2,0.2], [0.5,0.7], 'k-', linewidth=2)
ax.text(0.2, 0.75, 'AND', ha='center', fontsize=12)

# Dissonant Disjunction (OR)
ax.plot([0.5,0.7], [0.7,0.7], 'k-', linewidth=2)
ax.plot([0.6,0.6], [0.5,0.7], 'k-', linewidth=2)
ax.text(0.6, 0.75, 'OR', ha='center', fontsize=12)

# Negation (NOT)
ax.plot([0.8,0.9], [0.7,0.7], 'k-', linewidth=2)
ax.plot([0.85,0.85], [0.5,0.7], 'k-', linewidth=2)
ax.text(0.85, 0.75, 'NOT', ha='center', fontsize=12)

ax.text(0.2, 0.5, r'$r_1 \wedge r_2$', ha='center', fontsize=14)
ax.text(0.6, 0.5, r'$r_1 \vee r_2$', ha='center', fontsize=14)
ax.text(0.85, 0.5, r'$\neg r$', ha='center', fontsize=14)

ax.axis('off')
plt.title('Logical Connectives in Resonance Logic')
plt.show()
```
